```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Baca citra grayscale
img = cv2.imread('bunga tulip.jpg', cv2.IMREAD_GRAYSCALE)

# Pastikan citra berhasil dibaca
if img is None:
    raise ValueError("Gambar tidak ditemukan! Pastikan path file benar.")

# 2. Transformasi Negatif
negatif = 255 - img

# 3. Transformasi Logaritmik
c = 255 / np.log(1 + np.max(img))
log_transform = c * np.log(1 + img)
log_transform = np.array(log_transform, dtype=np.uint8)

# 4. Tampilkan hasil
plt.figure(figsize=(12, 10))

# Citra Asli
plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.hist(img.ravel(), bins=256, color='gray')
plt.title('Histogram Citra Asli')

# Citra Negatif
plt.subplot(3, 2, 3)
plt.imshow(negatif, cmap='gray')
plt.title('Citra Negatif')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.hist(negatif.ravel(), bins=256, color='gray')
plt.title('Histogram Citra Negatif')

# Citra Logaritmik
plt.subplot(3, 2, 5)
plt.imshow(log_transform, cmap='gray')
plt.title('Citra Logaritmik')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.hist(log_transform.ravel(), bins=256, color='gray')
plt.title('Histogram Citra Logaritmik')

plt.tight_layout()
plt.show()
```
