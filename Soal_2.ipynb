```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Baca citra grayscale
img = cv2.imread('bunga tulip.jpg', cv2.IMREAD_GRAYSCALE)

# Pastikan citra berhasil dibaca
if img is None:
    raise ValueError("Gambar tidak ditemukan! Pastikan path file benar.")

# 2. Definisikan kernel filter
# Low Pass Filter (LPF) - untuk mereduksi noise / blur
kernel_lpf = (1/9) * np.array([[1,1,1],
                               [1,1,1],
                               [1,1,1]])

# High Pass Filter (HPF) - untuk deteksi tepi
kernel_hpf = np.array([[-1,-1,-1],
                       [-1, 8,-1],
                       [-1,-1,-1]])

# Band Pass Filter (BPF) - untuk penajaman (sharpening)
kernel_bpf = np.array([[ 0,-1, 0],
                       [-1, 5,-1],
                       [ 0,-1, 0]])

# 3. Terapkan konvolusi menggunakan filter2D dari OpenCV
img_lpf = cv2.filter2D(img, -1, kernel_lpf)
img_hpf = cv2.filter2D(img, -1, kernel_hpf)
img_bpf = cv2.filter2D(img, -1, kernel_bpf)

# 4. Tampilkan hasil
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img_lpf, cmap='gray')
plt.title('Low Pass Filter (LPF) - Blur')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(img_hpf, cmap='gray')
plt.title('High Pass Filter (HPF) - Deteksi Tepi')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(img_bpf, cmap='gray')
plt.title('Band Pass Filter (BPF) - Penajaman')
plt.axis('off')

plt.tight_layout()
plt.show()
```
